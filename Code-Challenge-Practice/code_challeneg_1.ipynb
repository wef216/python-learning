```python
### MY SOLUTION:
def trans(str):
    transcode = []
    for letter in str:
        if letter.islower():
            decode = 96
            recode = ord(letter) - 96
            transcode.append(chr((recode + 2)%26 + 96))
        elif letter.isupper():
            recode = ord(letter) - 64
            transcode.append(chr((recode+2)%26 + 64))
        else:
            transcode.append(letter)
    new_str = "".join(new_letter for new_letter in transcode)
    print(new_str)

trans(inputs)
```
